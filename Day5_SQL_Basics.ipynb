```python
import sqlite3
import pandas as pd
#Create a temporary database memory
conn = sqlite3.connect(':memory:')
data ={
    "Shop" : ["Vibe Vault","Sola Mode","Aura Attire","Glow Theory"],
    "Sales": [45000,72000,65000,88000],
    "City": ["New York"," Los Angeles","Chicago","Ireland"]
}
df= pd.DataFrame(data)
df.to_sql("shops", conn,index = False, if_exists ="replace")
print("Table created suscesfully")
```

```python
result = pd.read_sql("SELECT * FROM shops", conn)
print(result)
```

```python
high_sales = pd.read_sql("SELECT * FROM shops WHERE Sales > 65000", conn)
print(high_sales)
```

```python
sorted_shops = pd.read_sql("SELECT * FROM shops ORDER BY Sales DESC", conn)
print(sorted_shops)
```
