```python
import pandas as pd
import numpy as np

# Create customers table
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'age': [45, 37, 50, 28, 35],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Chennai', 'Mumbai']
})

# Create transactions table
transactions = pd.DataFrame({
    'transaction_id': [101, 102, 103, 104, 105, 106],
    'customer_id': [1, 2, 1, 3, 2, 4],
    'amount': [100, 50, 75, 200, 120, 80],
    'class': [0, 0, 0, 1, 0, 0]  # 1 = fraud, 0 = legitimate
})

print("=" * 70)
print("DAY 10: SQL JOINs - LINKING FRAUD TO CUSTOMERS")
print("=" * 70)

print("\n📊 CUSTOMERS TABLE:")
print(customers)

print("\n💳 TRANSACTIONS TABLE:")
print(transactions)

print("\n🎯 GOAL: Connect transactions to customer details using JOINs")
```

```python
print("\n" + "=" * 70)
print("STEP 1: INNER JOIN - Only matching records")
print("=" * 70)

# INNER JOIN
inner_join = transactions.merge(customers, on='customer_id', how='inner')

print("\nResult (All transactions with customer details):")
print(inner_join)

print("\n🔍 What Happened:")
print(f"- Transactions: 6 rows")
print(f"- Customers: 5 rows")
print(f"- INNER JOIN result: {len(inner_join)} rows")
print(f"- All transactions had matching customers")

print("\n📝 INNER JOIN LOGIC:")
print("- Take transactions table (LEFT)")
print("- Match with customers table (RIGHT)")
print("- Show ONLY rows where customer_id exists in BOTH")
print("- Result: 6 rows (all matched)")
```

```python
print("\n" + "=" * 70)
print("STEP 2: LEFT JOIN - All from left + matching from right")
print("=" * 70)

# LEFT JOIN
left_join = customers.merge(transactions, on='customer_id', how='left')

print("\nResult (All customers + their transactions):")
print(left_join)

print("\n🔍 What Happened:")
print(f"- Customers table (left): 5 rows")
print(f"- Transactions table (right): 6 rows")
print(f"- LEFT JOIN result: {len(left_join)} rows")

print("\n📝 LEFT JOIN LOGIC:")
print("- Start with customers (LEFT table) = 5 customers")
print("- Match with transactions (RIGHT table)")
print("- Show ALL customers (even if no transactions)")
print("- NaN = No transaction for that customer")

print("\n🔎 Finding:")
customers_with_no_transactions = left_join[left_join['transaction_id'].isna()]['name'].unique()
print(f"Customers with NO transactions: {list(customers_with_no_transactions)}")
```

```python
print("\n" + "=" * 70)
print("STEP 3: FRAUD ANALYSIS - Link fraud to customer details")
print("=" * 70)

# INNER JOIN for analysis
fraud_data = transactions.merge(customers, on='customer_id', how='inner')

# Filter only fraudulent transactions
frauds = fraud_data[fraud_data['class'] == 1]

print("\nFraudulent Transactions:")
print(frauds[['customer_id', 'name', 'age', 'city', 'amount', 'class']])

print("\n🔍 KEY FINDINGS:")
print(f"Total frauds detected: {len(frauds)}")

if len(frauds) > 0:
    print(f"\nFraudster details:")
    for idx, row in frauds.iterrows():
        print(f"  - {row['name']} (Age {row['age']}, {row['city']}) stole €{row['amount']}")

print("\n💡 BUSINESS INSIGHT:")
print("Now we know WHO committed fraud!")
print("Before: 'Transaction 104 is fraud'")
print("After: 'Charlie (age 50, Bangalore) committed fraud for €200'")

print("\n🎯 ACTION ITEMS:")
print("✅ Block Charlie's card")
print("✅ Refund €200")
print("✅ Monitor other 50+ year old customers")
print("✅ Investigate Bangalore transactions")
```

```python
print("\n" + "=" * 70)
print("STEP 4: CUSTOMER RISK SCORE - Who might commit fraud?")
print("=" * 70)

# Join all data
all_data = transactions.merge(customers, on='customer_id', how='inner')

# Calculate risk by customer
customer_risk = all_data.groupby(['customer_id', 'name', 'age', 'city']).agg({
    'class': ['sum', 'count'],
    'amount': 'mean'
}).reset_index()

# Flatten column names
customer_risk.columns = ['customer_id', 'name', 'age', 'city', 'fraud_count', 'total_transactions', 'avg_amount']

# Calculate fraud rate
customer_risk['fraud_rate'] = (customer_risk['fraud_count'] / customer_risk['total_transactions'] * 100).round(1)

print("\nCustomer Risk Analysis:")
print(customer_risk)

print("\n⚠️ RISK ASSESSMENT:")
for idx, row in customer_risk.iterrows():
    if row['fraud_count'] > 0:
        fraud_rate = row['fraud_rate']
        print(f"🔴 HIGH RISK: {row['name']} (Age {row['age']}, {row['city']}) - {row['fraud_count']} fraud(s), {fraud_rate}% fraud rate")
    else:
        print(f"✅ LOW RISK: {row['name']} (Age {row['age']}, {row['city']}) - Clean record")

print("\n🎯 BUSINESS DECISIONS:")
print("High-risk customer actions:")
print("- 🚫 Block account? NO (legal issues)")
print("- 📊 Monitor closely? YES")
print("- 🔐 Require 2FA? YES")
print("- 💳 Increase limits? NO")
print("- ⚠️ Send security alerts? YES")

```

```python
import sqlite3

print("\n" + "=" * 70)
print("STEP 5: REAL SQL - Using SQLite (same logic)")
print("=" * 70)

# Create SQLite connection (in-memory database)
conn = sqlite3.connect(':memory:')

# Write tables to SQLite
customers.to_sql('customers', conn, index=False, if_exists='replace')
transactions.to_sql('transactions', conn, index=False, if_exists='replace')

# SQL INNER JOIN query
query = """
SELECT
    t.transaction_id,
    c.customer_id,
    c.name,
    c.age,
    c.city,
    t.amount,
    t.class
FROM transactions t
INNER JOIN customers c ON t.customer_id = c.customer_id
WHERE t.class = 1
"""

result = pd.read_sql(query, conn)

print("\n📊 SQL Query Result (Fraudulent Transactions):")
print(result)

print("\n📝 SQL QUERY BREAKDOWN:")
print("SELECT: Choose which columns to show")
print("  → transaction_id, name, age, city, amount, class")
print("\nFROM transactions t: Start with transactions table")
print("\nINNER JOIN customers c: Connect customers table")
print("\nON t.customer_id = c.customer_id: Match by customer_id")
print("\nWHERE t.class = 1: Filter only frauds")

print("\n💡 This is REAL SQL that banks use!")
print("Same result as Pandas, but using actual database query language")

conn.close()
```

```python
print("\n" + "=" * 70)
print("DAY 10 SUMMARY: JOINs & FRAUD ANALYSIS")
print("=" * 70)

print("\n🎓 WHAT YOU LEARNED:")
print("1. ✅ INNER JOIN: Only matching records")
print("2. ✅ LEFT JOIN: All from left + matching right")
print("3. ✅ How to link fraud to customer details")
print("4. ✅ How to calculate customer risk scores")
print("5. ✅ Real SQL queries vs Pandas")

print("\n📊 YOUR ANALYSIS RESULTS:")
print(f"- Total transactions analyzed: {len(transactions)}")
print(f"- Total customers: {len(customers)}")
print(f"- Frauds detected: {len(frauds)}")
print(f"- Fraud rate: {len(frauds)/len(transactions)*100:.1f}%")
print(f"- High-risk customers: {len(customer_risk[customer_risk['fraud_count']>0])}")
print(f"- Clean customers: {len(customer_risk[customer_risk['fraud_count']==0])}")

print("\n💰 BUSINESS VALUE:")
print("✅ Identified who committed fraud (Charlie)")
print("✅ Found fraud patterns by demographics")
print("✅ Scored customer risk levels")
print("✅ Made actionable recommendations")

print("\n🎯 KEY BUSINESS OUTCOME:")
print("Before JOINs: 'Transaction €200 is fraud'")
print("After JOINs: 'Charlie (age 50, Bangalore) committed €200 fraud'")
print("→ Can now take action!")

print("\n📈 NEXT DAYS (11-14):")
print("- Day 11: RIGHT JOIN & FULL OUTER JOIN")
print("- Day 12: More complex JOINs")
print("- Day 13: GROUP BY & Aggregations")
print("- Day 14: Advanced fraud queries")

print("\n✅ DAY 10 COMPLETE!")
print("=" * 70)
```

```python
print("\n" + "=" * 70)
print("BONUS: DEEP DIVE ANALYSIS")
print("=" * 70)

# Question 1: What's the fraud amount by customer age?
print("\n❓ Question: Does age affect fraud amount?")
age_fraud = fraud_data.groupby('age')[['amount', 'class']].agg({
    'amount': 'mean',
    'class': 'count'
}).reset_index()
age_fraud.columns = ['age', 'avg_amount', 'transaction_count']
print(age_fraud)

# Question 2: Which city has fraud?
print("\n❓ Question: Which cities have fraud?")
city_fraud = fraud_data[fraud_data['class'] == 1].groupby('city').size().reset_index(name='fraud_count')
print(city_fraud)

# Question 3: Customer spending pattern
print("\n❓ Question: Who spends the most?")
spending = all_data.groupby(['name', 'age']).agg({
    'amount': ['sum', 'mean', 'count']
}).reset_index()
spending.columns = ['name', 'age', 'total_spent', 'avg_amount', 'transaction_count']
print(spending.sort_values('total_spent', ascending=False))
```
