```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('memory :')
data = {
    "Shop" : ["Vibe Vault", "Sola Mode", "Aura Attire", "Glow Theory"],
    "Sales" : [45000, 72000, 65000 , 88000],
    "City" : ["New York", "LosAngles", "Chicago" , "Ireland"]
}
df = pd.DataFrame(data)
df.to_sql("shops",conn, index=False, if_exists ="replace")
print("Table ready !")
```

```python
result = pd.read_sql("SELECT COUNT(*)FROM shops",conn)
print(result)
```

```python
result = pd.read_sql ("""
SELECT
    SUM(Sales) AS Total_Sales,
    AVG(Sales) AS Average_Sales,
    MIN(Sales) AS Minimum_Sales,
    MAX(Sales) AS Highest_Sales
FROM shops
""" ,conn)
print(result)




```

```python
result = pd.read_sql("""
SELECT
     City,
     SUM(Sales)As Total_Sales
  FROM shops
  GROUP BY City
  """,conn)
print(result)
```
