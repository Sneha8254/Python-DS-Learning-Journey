```python
names = ["Vibe Vault", "Sola Mode", "Aura Attire", "Glow Theory"]
sales = [45000, 72000, 65000, 88000]

for i in range(len(names)):
  print(names[i], "made sale of",sales[i])

```

```python
#Find the best and wrost shop
best =  max(sales)
worst = min(sales)

print("Best sales:", best)
print("Worst sales:", worst)

#Total and average
total = sum(sales)
average = total / len(sales)

print("Total sales :", total)
print("Average sales :" , average)
```

```python
# Label each shop as high or low performer
for i in range(len(names)):
    if sales[i] > average :
      print(names[i], "is a High Performer")
    else :
      print(names[i],"is a Low Performer")
```
