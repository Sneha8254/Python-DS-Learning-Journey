```python
import pandas as pd

#Create your first data
data = {
    "Shop": ["Vibe Vault","Sola Mode","Aura Attire","Glow Thoery"],
    "Sales": [45000,72000,65000,88000],
    "City" : ["New York","Los Angeles","Chicago","Ireland"]
}
#Covert to DataFrame
df = pd.DataFrame(data)
#Display it
print(df)
```

```python
#How many rows and coloums?
print(df.shape)
#Column names
print(df.columns)
#First look at data
print(df.head())
#Quick Summary
print(df.describe())
```

```python
#Show only  high-performing shops (sales above 65000)
high = df[df["Sales"]>65000]
print ("highperformer:")
print(high)

#Sort shops by sales (Lowest to  highest)
sorted_df = df.sort_values("Sales")
print("\nSorted by Sales:")
print(sorted_df)
```

```python
#Add new column - tax on each shop
df["Tax"] = df["Sales"] *  0.8

#Add perfomance label
df["Performace"] = ["Low","High","Low","High"]

#show final table
print(df)
```
